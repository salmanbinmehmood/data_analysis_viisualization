```python
cn1_imm['total']=cn1_imm.sum(axis=1,numeric_only=True)


cn1_imm.sort_values(by='total',ascending=True,inplace=True)
df_top15=cn1_imm['total'].tail(15)
df_top15=df_top15[:-1]
```

```python
df_top15.plot(kind='barh',figsize=(12,12),color='g')
plt.title('top 15 immigrants in canada')
plt.xlabel('years')
plt.ylabel('immigrants')
for index,value in enumerate(df_top15):
    label=format(int(value),',')
#     print(label,index)
plt.annotate(label,xy=(value-60000,index-0.10),color='black')

plt.show()
```
