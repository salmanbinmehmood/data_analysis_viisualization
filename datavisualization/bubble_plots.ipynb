```python
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
```

```python
cn_imm=pd.read_excel("F:\python\pythonProject\pythonProject\datavisualization\canada_imm\canada_imm.xlsx")
```

```python
cn_imm.drop(['Type','DEV','AREA','REG','Coverage'],inplace=True,axis=1)
```

```python
cn_imm.rename (columns={'OdName':'Country','AreaName':'continent','RegName':'region'},inplace=True)

cn_imm.columns=list(map(str,cn_imm.columns))
cn_imm.set_index('Country',inplace=True)
cn_imm['Total']=cn_imm.sum(axis=1,numeric_only=True)
years=list(map(str,range(1980,2014)))
print('data dimensions',cn_imm.shape)
```

```python
cn_imm
```

```python
# df_total=pd.DataFrame(cn_imm[years].sum(axis=0))
# df_total.index=map(int,years)


df_ab=cn_imm[years].transpose()
df_ab.head()
```

```python
df_ab.index=map(int,df_ab.index)
df_ab.index.name='Years'
df_ab.reset_index(inplace=True)
df_ab
```

```python
nom_brz=(df_ab['Brazil']-df_ab['Brazil'].min())/(df_ab['Brazil'].max()-df_ab['Brazil'].min())

nom_arg=(df_ab['Argentina']-df_ab['Argentina'].min())/(df_ab['Argentina'].max()-df_ab['Argentina'].min())
nom_pk=(df_ab['Pakistan']-df_ab['Pakistan'].min())/(df_ab['Pakistan'].max()-df_ab['Pakistan'].min())

nom_pol=(df_ab['Poland']-df_ab['Poland'].min())/(df_ab['Poland'].max()-df_ab['Poland'].min())

# print(nom_arg)
# print(nom_brz)
```

```python
ax0=df_ab.plot(kind='scatter',
               x='Years',
           y='Brazil',
               figsize=(12,6),
               alpha=0.6,
          color='red',
           s=nom_brz*1000+10,
          xlim=(1975,2015))

ax1=df_ab.plot(kind='scatter',
               x='Years',
           y='Argentina',
               alpha=0.7,
          color='blue',
          s=nom_arg*1000+10,
          ax=ax0)

# ax2=df_ab.plot(kind='scatter',
#                x='Years',
#            y='Pakistan',
#                alpha=0.8,
#           color='green',
#           s=nom_pk*1000+10,
#           ax=ax0)

ax0.set_title('immigrants from btazil and argentina')
ax0.set_ylabel('immigrants')
ax0.legend(['Brazil','Argentina','Pakistan'],loc='upper left',fontsize='x-large')

# plt.show()
```

```python
ax0=df_ab.plot(kind='scatter',
               x='Years',
           y='Poland',
               figsize=(12,6),
               alpha=0.6,
          color='red',
           s=nom_pol*1000+10,
          xlim=(1975,2015))

ax1=df_ab.plot(kind='scatter',
               x='Years',
           y='Pakistan',
               alpha=0.7,
          color='blue',
          s=nom_pk*1000+10,
          ax=ax0)

# ax2=df_ab.plot(kind='scatter',
#                x='Years',
#            y='Pakistan',
#                alpha=0.8,
#           color='green',
#           s=nom_pk*1000+10,
#           ax=ax0)

ax0.set_title('immigrants from Poland and Pakistan')
ax0.set_ylabel('immigrants')
ax0.legend(['Poland','Pakistan'],loc='upper left',fontsize='x-large')

# plt.show()
```
